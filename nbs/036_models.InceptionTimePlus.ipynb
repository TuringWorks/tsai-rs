# InceptionTimePlus> Enhanced InceptionTime with Modern Improvements

InceptionTimePlus extends InceptionTime with:- Squeeze-and-excitation attention- Flexible activation functions- Improved regularization- Support for variable-length inputs

```python
import tsai_rsimport numpy as np
```

## Model Configuration

```python
# Load sample dataX_train, y_train, X_test, y_test = tsai_rs.get_UCR_data('ECG200')X_train = tsai_rs.ts_standardize(X_train)# InceptionTimePlus configurationconfig = tsai_rs.InceptionTimePlusConfig(    c_in=X_train.shape[1],    c_out=len(np.unique(y_train)),    seq_len=X_train.shape[2],    nf=32,    depth=6,    bottleneck=True,    residual=True)print(f"InceptionTimePlus configuration:")print(f"  Input channels: {config.c_in}")print(f"  Output classes: {config.c_out}")print(f"  Sequence length: {config.seq_len}")print(f"  Filters: {config.nf}")print(f"  Depth: {config.depth}")
```

## Example: Binary Classification

```python
# Configure for binary classificationX_train, y_train, X_test, y_test = tsai_rs.get_UCR_data('ECG200')X_train = tsai_rs.ts_standardize(X_train)X_test = tsai_rs.ts_standardize(X_test)config = tsai_rs.InceptionTimePlusConfig(    c_in=X_train.shape[1],    c_out=len(np.unique(y_train)),    seq_len=X_train.shape[2])learner = tsai_rs.LearnerConfig(    lr=3e-4,    epochs=100,    batch_size=64)print(f"Model ready for training on {len(X_train)} samples")
```

## Example: Multivariate Classification

```python
# Configure for multivariate classificationX_train, y_train, X_test, y_test = tsai_rs.get_UCR_data('LSST')X_train = tsai_rs.ts_standardize(X_train, by_sample=True)X_test = tsai_rs.ts_standardize(X_test, by_sample=True)config = tsai_rs.InceptionTimePlusConfig(    c_in=X_train.shape[1],    c_out=len(np.unique(y_train)),    seq_len=X_train.shape[2],    nf=32)print(f"Multivariate model:")print(f"  Variables: {config.c_in}")print(f"  Classes: {config.c_out}")
```

## Key Differences from InceptionTime1. **SE blocks**: Channel attention for better feature selection2. **Flexible kernels**: Customizable kernel sizes3. **Modern initialization**: Better weight initialization4. **Output heads**: Classification, regression, forecasting support
