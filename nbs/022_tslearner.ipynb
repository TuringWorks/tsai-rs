# TSLearner> High-level API for time series learning

```python
import tsai_rsimport numpy as np
```

## TSClassifier

```python
class TSClassifier:    """High-level classifier for time series."""        def __init__(self, X, y, model_config, learner_config):        self.X = X        self.y = y        self.model_config = model_config        self.learner_config = learner_config        self.is_fitted = False        def fit(self):        """Train the classifier."""        # Training would happen here via Rust        self.is_fitted = True        print(f"Training on {len(self.X)} samples")        print(f"  Epochs: {self.learner_config.epochs}")        print(f"  Batch size: {self.learner_config.batch_size}")        def predict(self, X):        """Make predictions."""        assert self.is_fitted, "Model not fitted"        return np.zeros(len(X), dtype=np.int64)        def predict_proba(self, X):        """Get prediction probabilities."""        assert self.is_fitted, "Model not fitted"        n_classes = self.model_config.c_out        return np.ones((len(X), n_classes)) / n_classes
```

```python
# Example usageX_train, y_train, X_test, y_test = tsai_rs.get_UCR_data('ECG200')X_train = tsai_rs.ts_standardize(X_train)model_cfg = tsai_rs.InceptionTimePlusConfig(    c_in=X_train.shape[1],    c_out=len(np.unique(y_train)),    seq_len=X_train.shape[2])learner_cfg = tsai_rs.LearnerConfig(lr=3e-4, epochs=100, batch_size=64)clf = TSClassifier(X_train, y_train, model_cfg, learner_cfg)clf.fit()
```
