## Sliding Window

```python
def sliding_window(data, window_size, horizon=1, stride=1):    """Create sliding windows from time series data.        Args:        data: 1D or 2D array of time series        window_size: Size of input window        horizon: Prediction horizon (steps ahead)        stride: Step size between windows        Returns:        X: Input windows [n_windows, channels, window_size]        y: Target values [n_windows, horizon]    """    data = np.asarray(data)    if data.ndim == 1:        data = data.reshape(1, -1)        n_channels, n_steps = data.shape    n_windows = (n_steps - window_size - horizon + 1) // stride + 1        X = np.zeros((n_windows, n_channels, window_size))    y = np.zeros((n_windows, horizon))        for i in range(n_windows):        start = i * stride        X[i] = data[:, start:start + window_size]        y[i] = data[0, start + window_size:start + window_size + horizon]        return X.astype(np.float32), y.astype(np.float32)
```
