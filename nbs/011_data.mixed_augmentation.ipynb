# Mixed Augmentation> Combined augmentation strategies

```python
import tsai_rsimport numpy as np
```

## Cutout Augmentation

```python
def cutout(X, cut_ratio=0.1):    """Apply cutout augmentation (randomly zero out segments)."""    X = np.asarray(X).copy()    n_samples, n_channels, seq_len = X.shape    cut_len = int(seq_len * cut_ratio)        for i in range(n_samples):        start = np.random.randint(0, seq_len - cut_len + 1)        X[i, :, start:start + cut_len] = 0        return X
```

## Cutmix Augmentation

```python
def cutmix(X1, X2, y1, y2, alpha=1.0):    """Apply cutmix augmentation between two samples."""    seq_len = X1.shape[-1]    lam = np.random.beta(alpha, alpha)    cut_len = int(seq_len * (1 - lam))    start = np.random.randint(0, seq_len - cut_len + 1)        X_mixed = X1.copy()    X_mixed[..., start:start + cut_len] = X2[..., start:start + cut_len]        lam_adjusted = 1 - cut_len / seq_len    y_mixed = lam_adjusted * y1 + (1 - lam_adjusted) * y2        return X_mixed, y_mixed
```
