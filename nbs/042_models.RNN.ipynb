# RNN> Recurrent Neural Networks for Time Series

This module provides LSTM and GRU based models for time series:- **LSTM** (Long Short-Term Memory)- **GRU** (Gated Recurrent Unit)Based on:- Hochreiter & Schmidhuber (1997) for LSTM- Cho et al. (2014) for GRU

```python
import tsai_rsimport numpy as np
```

## Model Configuration

```python
# Load sample dataX_train, y_train, X_test, y_test = tsai_rs.get_UCR_data('ECG200')X_train = tsai_rs.ts_standardize(X_train)# RNN configurationconfig = tsai_rs.RNNPlusConfig(    c_in=X_train.shape[1],    c_out=len(np.unique(y_train)),    seq_len=X_train.shape[2],    hidden_size=128,    n_layers=2,    rnn_type='LSTM',  # or 'GRU'    bidirectional=False,    dropout=0.1)print(f"RNN configuration:")print(f"  Hidden size: {config.hidden_size}")print(f"  Layers: {config.n_layers}")print(f"  Type: {config.rnn_type}")
```

## LSTM vs GRU- **LSTM**: Better for longer sequences, more parameters- **GRU**: Faster training, fewer parameters, often similar performance
