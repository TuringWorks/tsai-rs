# RNNPlus> Enhanced RNN with Modern Improvements

RNNPlus extends standard RNNs with:- Bidirectional support- Attention mechanisms- Residual connections- Flexible output heads

```python
import tsai_rsimport numpy as np
```

## Model Configuration

```python
# Load sample dataX_train, y_train, X_test, y_test = tsai_rs.get_UCR_data('ECG200')X_train = tsai_rs.ts_standardize(X_train)# RNNPlus configurationconfig = tsai_rs.RNNPlusConfig(    c_in=X_train.shape[1],    c_out=len(np.unique(y_train)),    seq_len=X_train.shape[2],    hidden_size=128,    n_layers=2,    rnn_type='LSTM',    bidirectional=True,    dropout=0.1)print(f"RNNPlus configuration:")print(f"  Hidden size: {config.hidden_size}")print(f"  Bidirectional: {config.bidirectional}")print(f"  Effective hidden: {config.hidden_size * (2 if config.bidirectional else 1)}")
```

## Bidirectional RNNsBidirectional RNNs process sequences in both directions, capturing:- Past context (forward pass)- Future context (backward pass)
