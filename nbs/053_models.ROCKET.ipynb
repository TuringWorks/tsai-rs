# ROCKET> RandOm Convolutional KErnel Transform

Based on:Dempster, A., Petitjean, F., & Webb, G. I. (2020). **ROCKET: exceptionally fast and accurate time series classification using random convolutional kernels**. Data Mining and Knowledge Discovery.Paper: https://arxiv.org/abs/1910.13051

```python
import tsai_rsimport numpy as np
```

## OverviewROCKET is a transform-based approach:1. Generate random convolutional kernels2. Apply kernels to extract features (max and proportion of positive values)3. Train a linear classifier on the featuresKey benefits:- Extremely fast training- State-of-the-art accuracy- No hyperparameter tuning required

```python
# Load sample dataX_train, y_train, X_test, y_test = tsai_rs.get_UCR_data('ECG200')X_train = tsai_rs.ts_standardize(X_train)X_test = tsai_rs.ts_standardize(X_test)print(f"Dataset: ECG200")print(f"Train shape: {X_train.shape}")print(f"Test shape: {X_test.shape}")
```

## Feature ExtractionFor each kernel, ROCKET extracts two features:- **Maximum value**: Captures the strongest activation- **Proportion of positive values (PPV)**: Captures pattern frequency

```python
# ROCKET typically uses 10,000 kernels# This results in 20,000 features (2 per kernel)n_kernels = 10000n_features = n_kernels * 2print(f"ROCKET features:")print(f"  Kernels: {n_kernels}")print(f"  Features per sample: {n_features}")
```
