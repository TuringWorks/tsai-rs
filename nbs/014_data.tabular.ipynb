# Tabular Data> Combining time series with tabular features

```python
import tsai_rsimport numpy as np
```

## Mixed Input Preparation

```python
def prepare_mixed_input(X_ts, X_tab):    """Prepare mixed time series and tabular input."""    n_samples = len(X_ts)    assert len(X_tab) == n_samples, "Sample count mismatch"        return {        'ts': X_ts.astype(np.float32),        'tab': X_tab.astype(np.float32)    }
```

```python
# Example with synthetic tabular dataX_train, _, _, _ = tsai_rs.get_UCR_data('ECG200')X_tab = np.random.randn(len(X_train), 10)  # 10 tabular featuresmixed = prepare_mixed_input(X_train, X_tab)print(f"Time series shape: {mixed['ts'].shape}")print(f"Tabular shape: {mixed['tab'].shape}")
```
