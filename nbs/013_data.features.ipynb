# Data Features> Feature extraction from time series

```python
import tsai_rsimport numpy as np
```

## Statistical Features

```python
def extract_features(X):    """Extract statistical features from time series."""    features = []        for sample in X:        sample_features = []        for channel in sample:            sample_features.extend([                np.mean(channel),                np.std(channel),                np.min(channel),                np.max(channel),                np.median(channel),                np.percentile(channel, 25),                np.percentile(channel, 75)            ])        features.append(sample_features)        return np.array(features)
```
