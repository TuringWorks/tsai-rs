# ResNetPlus> Enhanced ResNet for Time Series with Additional Features

ResNetPlus extends the standard ResNet with:- Squeeze-and-excitation blocks- Configurable activation functions- Dropout regularization- Custom output heads

```python
import tsai_rsimport numpy as np
```

## Model Configuration

```python
# Load sample dataX_train, y_train, X_test, y_test = tsai_rs.get_UCR_data('ECG200')X_train = tsai_rs.ts_standardize(X_train)# ResNetPlus configurationconfig = tsai_rs.ResNetPlusConfig(    c_in=X_train.shape[1],    c_out=len(np.unique(y_train)),    seq_len=X_train.shape[2],    nf=64)print(f"ResNetPlus for classification:")print(f"  Input: [{config.c_in}, {config.seq_len}]")print(f"  Output: {config.c_out} classes")
```
