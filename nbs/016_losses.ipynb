## Classification Losses

```python
def cross_entropy_loss(y_pred, y_true):    """Cross entropy loss for classification."""    y_pred = np.clip(y_pred, 1e-7, 1 - 1e-7)    n_classes = y_pred.shape[1]    y_one_hot = np.eye(n_classes)[y_true]    return -np.mean(np.sum(y_one_hot * np.log(y_pred), axis=1))def focal_loss(y_pred, y_true, gamma=2.0, alpha=0.25):    """Focal loss for imbalanced classification."""    y_pred = np.clip(y_pred, 1e-7, 1 - 1e-7)    n_classes = y_pred.shape[1]    y_one_hot = np.eye(n_classes)[y_true]    p_t = np.sum(y_one_hot * y_pred, axis=1)    return -np.mean(alpha * (1 - p_t) ** gamma * np.log(p_t))
```

## Regression Losses

```python
def mse_loss(y_pred, y_true):    """Mean squared error loss."""    return np.mean((y_pred - y_true) ** 2)def huber_loss(y_pred, y_true, delta=1.0):    """Huber loss for robust regression."""    error = y_pred - y_true    abs_error = np.abs(error)    quadratic = np.minimum(abs_error, delta)    linear = abs_error - quadratic    return np.mean(0.5 * quadratic ** 2 + delta * linear)
```
