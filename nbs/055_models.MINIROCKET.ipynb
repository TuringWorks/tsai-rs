# MiniRocket> A Very Fast (Almost) Deterministic Transform for Time Series Classification

Based on:Dempster, A., Schmidt, D. F., & Webb, G. I. (2021). **MINIROCKET: A Very Fast (Almost) Deterministic Transform for Time Series Classification**. KDD 2021.Paper: https://arxiv.org/abs/2102.00457Key improvements over ROCKET:- 75x faster- Nearly deterministic (fixed kernels)- Similar accuracy

## Key Differences from ROCKET| Feature | ROCKET | MiniRocket ||---------|--------|------------|| Kernels | Random | Fixed (84 kernels) || Dilations | Random | Fixed || Features | Max + PPV | PPV only || Speed | Fast | 75x Faster |

```python
# MiniRocket uses fixed kernels of length 9# with 84 unique kernel patterns# Combined with various dilations for multi-scaleprint("MiniRocket kernel info:")print("  Kernel length: 9")print("  Unique kernels: 84")print("  Feature type: PPV (proportion of positive values)")
```
