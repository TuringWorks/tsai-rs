# External Data> Functions to download and load external time series datasets

This module provides functions to download and load popular time series datasets including:- **UCR/UEA Archive**: 128 univariate and 30 multivariate classification datasets- **Monash Repository**: Regression and forecasting datasets

```python
import tsai_rsimport numpy as np
```

## UCR/UEA Classification DatasetsThe UCR/UEA archive contains the largest collection of time series classification datasets.

```python
# List of available UCR univariate datasets (sample)UCR_DATASETS = [    'ECG200', 'GunPoint', 'Wafer', 'FaceAll', 'Fish', 'Beef',    'Coffee', 'OliveOil', 'ArrowHead', 'CBF', 'Trace',    'SyntheticControl', 'TwoPatterns', 'FordA', 'FordB']print(f"Sample UCR datasets: {UCR_DATASETS}")
```

## Loading UCR Data with tsai_rs

```python
# Load a univariate classification datasetX_train, y_train, X_test, y_test = tsai_rs.get_UCR_data('ECG200')print(f"Dataset: ECG200")print(f"X_train shape: {X_train.shape}")print(f"X_test shape: {X_test.shape}")print(f"Classes: {np.unique(y_train)}")print(f"Train class distribution: {dict(zip(*np.unique(y_train, return_counts=True)))}")
```

```python
# Load a multivariate classification datasetX_train, y_train, X_test, y_test = tsai_rs.get_UCR_data('NATOPS')print(f"\nDataset: NATOPS (multivariate)")print(f"X_train shape: {X_train.shape}")print(f"X_test shape: {X_test.shape}")print(f"Number of channels: {X_train.shape[1]}")print(f"Sequence length: {X_train.shape[2]}")
```

## Data Formattsai_rs uses the standard format for time series data:**X: [n_samples, n_channels, seq_len]**- n_samples: Number of time series samples- n_channels: Number of variables/features- seq_len: Length of each time series**y: [n_samples]**- Labels for classification tasks

```python
def inspect_dataset(name):    """Load and display dataset info."""    X_train, y_train, X_test, y_test = tsai_rs.get_UCR_data(name)        print(f"Dataset: {name}")    print(f"  Train samples: {X_train.shape[0]}")    print(f"  Test samples: {X_test.shape[0]}")    print(f"  Channels: {X_train.shape[1]}")    print(f"  Sequence length: {X_train.shape[2]}")    print(f"  Classes: {len(np.unique(y_train))}")    print()
```

```python
# Inspect a few datasetsfor dsid in ['ECG200', 'GunPoint', 'Wafer']:    inspect_dataset(dsid)
```
