## Array Shape UtilitiesFunctions for converting arrays between different dimensionalities.

```python
def to3d(arr):    """Convert array to 3D [samples x channels x length]."""    arr = np.asarray(arr)    if arr.ndim == 3:        return arr    elif arr.ndim == 1:        return arr[None, None, :]    elif arr.ndim == 2:        return arr[:, None, :]    else:        raise ValueError(f"Cannot convert {arr.ndim}D array to 3D")def to2d(arr):    """Convert array to 2D [samples x length]."""    arr = np.asarray(arr)    if arr.ndim == 2:        return arr    elif arr.ndim == 1:        return arr[None, :]    elif arr.ndim == 3:        return arr[:, 0, :]    else:        raise ValueError(f"Cannot convert {arr.ndim}D array to 2D")def to1d(arr):    """Convert array to 1D [length]."""    arr = np.asarray(arr)    if arr.ndim == 1:        return arr    elif arr.ndim == 2:        return arr[0, :]    elif arr.ndim == 3:        return arr[0, 0, :]    else:        raise ValueError(f"Cannot convert {arr.ndim}D array to 1D")
```

## Random Utilities

```python
def set_seed(seed):    """Set random seed for reproducibility."""    np.random.seed(seed)def random_split(n, val_size=0.2, seed=None):    """Create random train/validation split indices."""    if seed is not None:        np.random.seed(seed)    indices = np.random.permutation(n)    val_n = int(n * val_size)    return indices[val_n:], indices[:val_n]
```

## Data Validation

```python
def check_data(X, y=None):    """Validate time series data format."""    X = np.asarray(X)        if X.ndim != 3:        raise ValueError(f"X must be 3D [samples x channels x length], got {X.ndim}D")        if np.isnan(X).any():        print("Warning: X contains NaN values")        if y is not None:        y = np.asarray(y)        if len(X) != len(y):            raise ValueError(f"X and y must have same length: {len(X)} vs {len(y)}")        print(f"Data shape: {X.shape}")    print(f"Data dtype: {X.dtype}")    print(f"Value range: [{X.min():.4f}, {X.max():.4f}]")        return X, y
```

## Batch Utilities

```python
def batch_iterator(X, y=None, batch_size=32, shuffle=True):    """Iterate over data in batches."""    n = len(X)    indices = np.arange(n)        if shuffle:        np.random.shuffle(indices)        for start in range(0, n, batch_size):        end = min(start + batch_size, n)        batch_idx = indices[start:end]                if y is not None:            yield X[batch_idx], y[batch_idx]        else:            yield X[batch_idx]
```
