# ResNet> Residual Network for Time Series Classification

This is a Rust implementation of ResNet for time series, based on:Wang, Z., Yan, W., & Oates, T. (2017). **Time series classification from scratch with deep neural networks: A strong baseline**. IJCNN 2017.The original paper: https://arxiv.org/abs/1611.06455

```python
import tsai_rsimport numpy as np
```

## ArchitectureResNet uses residual connections to enable training of deeper networks:1. Initial convolution layer2. Three residual blocks with increasing filter sizes3. Global average pooling4. Fully connected output layer

```python
# Load sample dataX_train, y_train, X_test, y_test = tsai_rs.get_UCR_data('ECG200')X_train = tsai_rs.ts_standardize(X_train)X_test = tsai_rs.ts_standardize(X_test)print(f"Input shape: {X_train.shape}")print(f"Classes: {np.unique(y_train)}")
```

## Model Configuration

```python
# ResNet configurationconfig = tsai_rs.ResNetPlusConfig(    c_in=X_train.shape[1],      # Number of input channels    c_out=len(np.unique(y_train)),  # Number of output classes    seq_len=X_train.shape[2],   # Sequence length    nf=64                       # Number of filters)print(f"ResNet configuration:")print(f"  Input channels: {config.c_in}")print(f"  Output classes: {config.c_out}")print(f"  Sequence length: {config.seq_len}")print(f"  Filters: {config.nf}")
```

## Training Configuration

```python
# Configure traininglearner_config = tsai_rs.LearnerConfig(    lr=1e-3,    epochs=50,    batch_size=64,    optimizer='AdamW',    weight_decay=1e-2)print(f"Training configuration:")print(f"  Learning rate: {learner_config.lr}")print(f"  Epochs: {learner_config.epochs}")print(f"  Batch size: {learner_config.batch_size}")
```

## Key Features- **Residual connections**: Enable gradient flow in deep networks- **Batch normalization**: Stabilize training- **1D convolutions**: Designed for time series data- **Global average pooling**: Reduces overfitting
