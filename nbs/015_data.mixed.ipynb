## Multi-Modal Data

```python
class MultiModalDataset:    """Dataset for multiple input modalities."""        def __init__(self, modalities, y=None):        """        Args:            modalities: Dict of {name: data}            y: Labels        """        self.modalities = modalities        self.y = y                # Verify all have same length        lengths = [len(v) for v in modalities.values()]        assert len(set(lengths)) == 1, "All modalities must have same length"        def __len__(self):        return len(list(self.modalities.values())[0])        def __getitem__(self, idx):        batch = {k: v[idx] for k, v in self.modalities.items()}        if self.y is not None:            return batch, self.y[idx]        return batch
```
