## UCR Archive Metadataset

```python
# UCR univariate datasets by categoryUCR_DATASETS_BY_CATEGORY = {    'ECG': ['ECG200', 'ECG5000', 'ECGFiveDays', 'TwoLeadECG'],    'Motion': ['GunPoint', 'Trace', 'Plane', 'ArrowHead'],    'Sensor': ['Wafer', 'FordA', 'FordB', 'Computers'],    'Image': ['FaceAll', 'FaceFour', 'FacesUCR', 'Fish'],    'Spectro': ['Beef', 'Coffee', 'OliveOil', 'Wine', 'Meat']}for category, datasets in UCR_DATASETS_BY_CATEGORY.items():    print(f"{category}: {len(datasets)} datasets")
```

```python
def load_multiple_datasets(dataset_names):    """Load multiple UCR datasets."""    results = {}    for name in dataset_names:        try:            X_train, y_train, X_test, y_test = tsai_rs.get_UCR_data(name)            results[name] = {                'X_train': X_train, 'y_train': y_train,                'X_test': X_test, 'y_test': y_test            }            print(f"Loaded {name}: {X_train.shape}")        except Exception as e:            print(f"Failed to load {name}: {e}")    return results
```
