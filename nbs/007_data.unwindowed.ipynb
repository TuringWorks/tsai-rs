## Variable Length Handling

```python
def pad_sequences(sequences, maxlen=None, padding='post', value=0.0):    """Pad variable-length sequences to same length.        Args:        sequences: List of 2D arrays [channels, length]        maxlen: Target length (default: max sequence length)        padding: 'pre' or 'post'        value: Padding value        Returns:        3D array [n_samples, channels, maxlen]    """    if maxlen is None:        maxlen = max(s.shape[-1] for s in sequences)        n_samples = len(sequences)    n_channels = sequences[0].shape[0]        padded = np.full((n_samples, n_channels, maxlen), value, dtype=np.float32)        for i, seq in enumerate(sequences):        length = seq.shape[-1]        if padding == 'post':            padded[i, :, :length] = seq        else:  # pre            padded[i, :, -length:] = seq        return padded
```
