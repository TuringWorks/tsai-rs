# TSTPlus> Enhanced Time Series Transformer

TSTPlus extends TST with:- Learnable positional encodings- Pre-norm vs post-norm options- Flexible attention patterns- Multiple output heads

```python
import tsai_rsimport numpy as np
```

## Model Configuration

```python
# Load sample dataX_train, y_train, X_test, y_test = tsai_rs.get_UCR_data('ECG200')X_train = tsai_rs.ts_standardize(X_train)# TSTPlus configurationconfig = tsai_rs.TSTConfig(    c_in=X_train.shape[1],    c_out=len(np.unique(y_train)),    seq_len=X_train.shape[2],    n_heads=8,    d_model=128,    d_ff=256,    n_layers=3,    dropout=0.1)print(f"TSTPlus configuration:")print(f"  Input: [{config.c_in}, {config.seq_len}]")print(f"  Output: {config.c_out}")print(f"  Model dim: {config.d_model}")print(f"  Heads: {config.n_heads}")print(f"  Layers: {config.n_layers}")
```

## Key Features- **Multi-head attention**: Captures different temporal patterns- **Feed-forward layers**: Non-linear transformations- **Layer normalization**: Stabilizes training- **Residual connections**: Enables deeper networks

```python
# Training configurationlearner = tsai_rs.LearnerConfig(    lr=1e-4,    epochs=100,    batch_size=64,    optimizer='AdamW',    weight_decay=1e-2)scheduler = tsai_rs.OneCycleLR(    max_lr=1e-4,    epochs=100,    pct_start=0.1)print("Training configuration set")
```
