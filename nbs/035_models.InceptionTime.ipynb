# InceptionTime> Ensemble of Deep Convolutional Neural Networks for Time Series Classification

InceptionTime is based on:Fawaz, H. I., et al. (2019). **InceptionTime: Finding AlexNet for Time Series Classification**. arXiv preprint arXiv:1909.04939.Key innovation: Multi-scale convolutional filters inspired by the Inception-v4 architecture.

```python
import tsai_rsimport numpy as np
```

## ArchitectureInceptionTime consists of:1. **Inception modules**: Multiple parallel convolutions with different kernel sizes2. **Bottleneck layers**: Dimensionality reduction3. **Residual connections**: Every 3 modules4. **Global average pooling**: Before the output layer

```python
# Load sample dataX_train, y_train, X_test, y_test = tsai_rs.get_UCR_data('ECG200')X_train = tsai_rs.ts_standardize(X_train)print(f"Dataset: ECG200")print(f"Shape: {X_train.shape}")print(f"Classes: {len(np.unique(y_train))}"))
```

## Model Configuration

```python
# InceptionTime configuration (basic)config = tsai_rs.InceptionTimePlusConfig(    c_in=X_train.shape[1],    c_out=len(np.unique(y_train)),    seq_len=X_train.shape[2],    nf=32,           # Number of filters per branch    depth=6,         # Number of inception modules    bottleneck=True, # Use bottleneck layers    residual=True    # Use residual connections)print(f"InceptionTime configuration:")print(f"  Filters: {config.nf}")print(f"  Depth: {config.depth}")
```
